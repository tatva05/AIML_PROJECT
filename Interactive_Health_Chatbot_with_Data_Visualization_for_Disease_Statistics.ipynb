```python
!pip install python-aiml

from google.colab import files
uploaded = files.upload()

import aiml

# Create a Kernel object
kernel = aiml.Kernel()

# Load the AIML file
aiml_filename = list(uploaded.keys())[0]  # Get the name of the uploaded file
kernel.learn(aiml_filename)

# Define a function to get responses from the chatbot
def get_response(user_input):
    return kernel.respond(user_input)

# Test the chatbot
test_input = "WHAT IS CANCER"
response = get_response(test_input)
print("Chatbot Response:", response)

import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {
    'Disease': ['Flu', 'Diabetes', 'Hypertension'],
    'Cases': [150, 200, 180]
}
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(8, 5))
plt.bar(df['Disease'], df['Cases'], color='skyblue')
plt.xlabel('Disease')
plt.ylabel('Number of Cases')
plt.title('Disease Statistics')
plt.show()
```
